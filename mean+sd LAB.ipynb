```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df1 = pd.read_excel("acetate.xlsx", sheet_name = "Reflectance")
df1
# df2 = pd.read_excel("7_7exp2.xlsx", sheet_name = "SR LAB")
```

```python
l = df1["L*"]
a = df1["a*"]
b = df1["b*"]
C = df1["C*"]
h = df1["h"]
lablist = [l, a, b, C, h]
lablabel = ["L", "a", "b", "C", "h"]

# leaves = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L"]
leaves = ["1G", "2G", "3G", "4G", "5G", "1G4Y", "2G3Y", "3G2Y", "4G1Y", "1G4O", "2G3O", "3G2O", "4G1O"]
stats = pd.DataFrame([0], columns = ["Remove"])

for lab, lablabel in zip(lablist, lablabel):
    i = 0
    means = []
    stdev = []
    labels = []
    for leafname in leaves:
        for number in [1,2,3,4]:
            means.append(np.mean(lab[i:i+3]))
            stdev.append(np.std(lab[i:i+3]))
            labels.append("%s%s"%(leafname, number)) #change letter here for experimental identifier
            i+=3
    dataframe = pd.DataFrame(list(zip(means,stdev)), columns = ["%s Mean"%lablabel, "%s stdev"%lablabel])
    stats = pd.concat([stats, dataframe], axis=1)

labelframe = pd.DataFrame(labels, columns = ["Leaf"])
stats = pd.concat([stats, labelframe], axis=1)
stats = stats.drop(["Remove"], axis = 1)



# #SR
# l_sr = df2["L*"]
# a_sr = df2["a*"]
# b_sr = df2["b*"]
# c_sr = df2["C*"]
# h_sr = df2["h"]
# lablist = [l_sr, a_sr, b_sr, c_sr, h_sr]
# lablabel = ["l", "a", "b", "c", "h"]

# leaves = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L"]
# stats_sr = pd.DataFrame([0], columns = ["Remove"])

# for lab, lablabel in zip(lablist, lablabel):
#     i = 0
#     means = []
#     stdev = []
#     labels = []
#     for leafname in leaves:
#         for number in [1,2,3,4]:
#             means.append(np.mean(lab[i:i+3]))
#             stdev.append(np.std(lab[i:i+3]))
#             labels.append("B%s%s"%(leafname, number)) #change letter
#             i+=3
#     dataframe = pd.DataFrame(list(zip(means,stdev)), columns = ["%s Mean"%lablabel, "%s stdev"%lablabel])
#     stats_sr = pd.concat([stats_sr, dataframe], axis=1)

# labelframe = pd.DataFrame(labels, columns = ["Leaf"])
# stats_sr = pd.concat([stats_sr, labelframe], axis=1)
# stats_sr = stats_sr.drop(["Remove"], axis = 1)
```
