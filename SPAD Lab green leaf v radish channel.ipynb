```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

filename = "7_7exp2"

df1 = pd.read_excel("7_7exp2.xlsx", sheet_name = "LAB")

df2 = pd.read_excel("7_7exp2.xlsx", sheet_name = "GL LAB")
```

```python
# Colourworker
leafname_df = df1["Comment"]
l_df = df1["L*"]
a_df = df1["a*"]
b_df = df1["b*"]
c_df = df1["C*"]
h_df = df1["h"]
spad_df = df1["SPAD"]
stdev_df = df1["stdev"]

leafname, l, a, b, c, h, spad, stdev = [], [], [], [], [], [], [], []
for i in range(len(a_df)):
    leafname.append(leafname_df[i])
    l.append(float(l_df[i]))
    a.append(float(a_df[i]))
    b.append(float(b_df[i]))
    c.append(float(c_df[i]))
    h.append(float(h_df[i]))
    spad.append(float(spad_df[i]))
    stdev.append(float(stdev_df[i]))

# specradio 
leafname2_df = df2["Comment"]
l2_df = df2["L*"]
a2_df = df2["a*"]
b2_df = df2["b*"]
c2_df = df2["C*"]
h2_df = df2["h"]
spad2_df = df2["SPAD"]

leafname_s, l_s, a_s, b_s, c_s, h_s, spad_s  = [], [], [], [], [], [], []
for i in range(len(a2_df)):
    leafname_s.append(leafname2_df[i])
    l_s.append(float(l2_df[i]))
    a_s.append(float(a2_df[i]))
    b_s.append(float(b2_df[i]))
    c_s.append(float(c2_df[i]))
    h_s.append(float(h2_df[i]))
    spad_s.append(float(spad2_df[i]))

nitrogen = ["AG", "AH", "BG", "BH","AI", "AJ", "BI", "BJ", "AK", "AL", "BK", "BL"]
water = ["AA", "AB", "BA", "BB", "AC", "AD", "BC", "BD", "AE", "AF", "BE", "BF"]
xlabs = [l, a, b, c, h]
xlabs_s = [l_s, a_s, b_s, c_s, h_s]
x_labels = ["L*", "a*", "b*", "C*", "h"]
file_label = ["L", "a", "b", "C", "h"]

for lab in range(len(xlabs)):
    colors = []
    labels = []
    indexes = []
    colors_s = []
    labels_s = []
    indexes_s = []
    for leaf in leafname:
        for n in nitrogen:
            if n in leaf:
                colors.append("#e68870")
                labels.append("Radish channel - nitrogen plants")
                indexes.append(leafname.index(leaf))
        for n in water:
                if n in leaf:
                    colors.append("#0296fa")
                    labels.append("Radish channel - water only plants")
                    indexes.append(leafname.index(leaf))
    for leaf in leafname_s:
        for n in nitrogen:
            if n in leaf:
                colors_s.append("#e68870")
                labels_s.append("Green leaf channel - nitrogen plants")
                indexes_s.append(leafname_s.index(leaf))
        for n in water:
            if n in leaf:
                colors_s.append("#0296fa")
                labels_s.append("Green leaf channel - water only plants")
                indexes_s.append(leafname_s.index(leaf))


    plt.figure(dpi = 600)
    for i in range(len(indexes)):
        plt.scatter((xlabs[lab])[indexes[i]], spad[indexes[i]], color = colors[i], label = labels[i], s = 19, edgecolors = '#454545', linewidths = 0.8)
    for i in range(len(indexes_s)):
        plt.scatter((xlabs_s[lab])[indexes_s[i]], spad_s[indexes_s[i]], color = colors_s[i], alpha = 0.5, label = labels_s[i], s = 19, edgecolors = None, linewidths = 0.8, )
    #
    handles, labels_d = plt.gca().get_legend_handles_labels()
    by_label = dict(zip(labels_d, handles))
    plt.legend(by_label.values(), by_label.keys(), fontsize = 8)
    #
    plt.title("SPAD Against %s for Experiment 2 plants: Green Leaf vs Radish Channel"%x_labels[lab])
    plt.xlabel("%s"%x_labels[lab])
    plt.ylabel("SPAD")
    plt.savefig("%s_N+W GL_RADISH_%s"%(filename, file_label[lab]))
    plt.show()
```
